# Problem 1: 

Create a program that attempts to read a file named data.txt. If the file does not exist, the program should not crash. Instead, it should catch the error and display a meaningful message to inform the user that the file was not found. This ensures the program gracefully handles missing files without interrupting its flow.

```python
try:
    with open("data.txt", "r") as file:
        content = file.read()
        print("File content:")
        print(content)
except FileNotFoundError:
    print("The file 'data.txt' was not found.")
```

## Problem 2: 

Write a program that takes two numbers as input from the user and performs division. The program should handle two potential errors: ZeroDivisionError and ValueError. If no error occurs, the program should calculate and display the result of the division.

```python
try:
    num1 = float(input("Enter the numerator: "))
    num2 = float(input("Enter the denominator: "))
    result = num1 / num2
except ZeroDivisionError:
    print("Division by zero is not allowed.")
except ValueError:
    print("Invalid input. Please enter numeric values.")
else:
    print("Success!", result)
```
